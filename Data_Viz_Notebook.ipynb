```python
# Making plots
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
def moving_average(array, window_size):
    """
    Calculates the moving average of a given array, using a specified window size efficiently.
    """
    # Check if window_size is greater than the length of the array
    if window_size > len(array):
        raise ValueError("window_size must be less than or equal to the length of the array.")
    
    # Calculate the cumulative sum of the array
    cum_sum = [0] * (len(array) + 1)
    for i in range(len(array)):
        cum_sum[i + 1] = cum_sum[i] + array[i]
        
    # Calculate the moving averages using the cumulative sum
    moving_avg = [(cum_sum[i + window_size] - cum_sum[i]) / window_size for i in range(len(array) - window_size + 1)]
    
    return moving_avg
```

```python
# Obtaining all the averages:
# 3 Grids:
her = np.load('grid_results/exp_grid_her_n3.npy')
sre = np.load('grid_results/exp_grid_sre_n3.npy')
re = np.load('grid_results/exp_grid_re_n3.npy')
sre_nc = np.load('grid_results/exp_grid_srenc_n3.npy')
sre_nitr = np.load('grid_results/exp_grid_srenitr_n3.npy')

window_size = 10000

ma_arr1 = moving_average(her, window_size)
ma_arr2 = moving_average(re, window_size)
ma_arr3 = moving_average(sre_nc, window_size)
ma_arr4 = moving_average(sre_nitr, window_size)
ma_arr5 = moving_average(sre, window_size)

print('her',ma_arr1)
print('re',ma_arr3)
```

```python
# 3x3 grid
import numpy as np
import matplotlib.pyplot as plt

# Load data
her = np.load('grid_results/exp_grid_her_n3.npy')
sre = np.load('grid_results/exp_grid_sre_n3.npy')
re = np.load('grid_results/exp_grid_re_n3.npy')
sre_nc = np.load('grid_results/exp_grid_srenc_n3.npy')
sre_nitr = np.load('grid_results/exp_grid_srenitr_n3.npy')

def moving_avg_std(data, window_size):
    """Compute moving average and standard deviation using cumulative sums."""
    cumsum = np.cumsum(data)
    cumsum2 = np.cumsum(data**2)
    
    moving_sum = cumsum[window_size:] - cumsum[:-window_size]
    moving_sum2 = cumsum2[window_size:] - cumsum2[:-window_size]
    
    moving_avg = moving_sum / window_size
    moving_var = (moving_sum2 - (moving_sum**2) / window_size) / window_size
    moving_std = np.sqrt(moving_var)
    
    return moving_avg, moving_std

# Set plotting parameters
window_size = 1000
indices = range(0, len(her) - window_size) # Adjust indices for the moving average
global_marker_frequency = 1000
line_thickness = 1.5
reduction_factor = 0

# Compute moving averages and standard deviations
ma_her, std_her = moving_avg_std(her, window_size)
ma_re, std_re = moving_avg_std(re, window_size)
ma_sre_nc, std_sre_nc = moving_avg_std(sre_nc, window_size)
ma_sre_nitr, std_sre_nitr = moving_avg_std(sre_nitr, window_size)
ma_sre, std_sre = moving_avg_std(sre, window_size)

# Plotting
plt.figure(figsize=(10, 6))

# For HER
plt.plot(indices, ma_her, label='HER', color='#2196F3', marker='o', linewidth=line_thickness, markevery=global_marker_frequency)
plt.fill_between(indices, ma_her - reduction_factor*std_her, ma_her + reduction_factor*std_her, color='#2196F3', alpha=0.2)

# For RE
plt.plot(indices, ma_re, label='RE', color='#4CAF50', marker='x', linewidth=line_thickness, markevery=global_marker_frequency)
plt.fill_between(indices, ma_re - reduction_factor*std_re, ma_re + reduction_factor*std_re, color='#4CAF50', alpha=0.2)

# For SRE_NC
plt.plot(indices, ma_sre_nc, label='SRE_NC', color='#F44336', marker='s', linewidth=line_thickness, markevery=global_marker_frequency)
plt.fill_between(indices, ma_sre_nc - reduction_factor*std_sre_nc, ma_sre_nc + reduction_factor*std_sre_nc, color='#F44336', alpha=0.2)

# For SRE_NITR
plt.plot(indices, ma_sre_nitr, label='SRE_NITR', color='#9C27B0', marker='^', linewidth=line_thickness, markevery=global_marker_frequency)
plt.fill_between(indices, ma_sre_nitr - reduction_factor*std_sre_nitr, ma_sre_nitr + reduction_factor*std_sre_nitr, color='#9C27B0', alpha=0.2)

# For SRE
plt.plot(indices, ma_sre, label='SRE', color='#FF9800', marker='d', linewidth=line_thickness, markevery=global_marker_frequency)
plt.fill_between(indices, ma_sre - reduction_factor*std_sre, ma_sre + reduction_factor*std_sre, color='#FF9800', alpha=0.2)

# Define text sizes
title_size = 20
label_size = 20
tick_size = 20
legend_size = 15

# Increase tick label size
plt.xticks(fontsize=tick_size)
plt.yticks(fontsize=tick_size)


# Adding title, labels, and legend
plt.title('3x3 Grid Results',fontsize = title_size)
plt.xlabel('# of Episodes',fontsize = label_size)
plt.ylabel('Average Success Rate',fontsize = label_size )
plt.legend(loc='upper left',fontsize = legend_size)  # Shift legend to upper left for better visibility

# Displaying the plot
plt.savefig("3grid.png", format="png", dpi=300)  # Save the plot with 300 dpi resolution
plt.show()
```

```python
# 15x15 grid
import numpy as np
import matplotlib.pyplot as plt

# Load data
her = np.load('grid_results/exp_grid_her_n15.npy')
sre = np.load('grid_results/exp_grid_sre_n15.npy')
re = np.load('grid_results/exp_grid_re_n15.npy')
sre_nc = np.load('grid_results/exp_grid_srenc_n15.npy')
sre_nitr = np.load('grid_results/exp_grid_srenitr_n15.npy')

def moving_avg_std(data, window_size):
    """Compute moving average and standard deviation using cumulative sums."""
    cumsum = np.cumsum(data)
    cumsum2 = np.cumsum(data**2)
    
    moving_sum = cumsum[window_size:] - cumsum[:-window_size]
    moving_sum2 = cumsum2[window_size:] - cumsum2[:-window_size]
    
    moving_avg = moving_sum / window_size
    moving_var = (moving_sum2 - (moving_sum**2) / window_size) / window_size
    moving_std = np.sqrt(moving_var)
    
    return moving_avg, moving_std

# Set plotting parameters
window_size = 1000
indices = range(0, len(her) - window_size) # Adjust indices for the moving average
global_marker_frequency = 1000
line_thickness = 1.5
reduction_factor = 0

# Compute moving averages and standard deviations
ma_her, std_her = moving_avg_std(her, window_size)
ma_re, std_re = moving_avg_std(re, window_size)
ma_sre_nc, std_sre_nc = moving_avg_std(sre_nc, window_size)
ma_sre_nitr, std_sre_nitr = moving_avg_std(sre_nitr, window_size)
ma_sre, std_sre = moving_avg_std(sre, window_size)

# Plotting
plt.figure(figsize=(10, 6))

# For HER
plt.plot(indices, ma_her, label='HER', color='#2196F3', marker='o', linewidth=line_thickness, markevery=global_marker_frequency)
plt.fill_between(indices, ma_her - reduction_factor*std_her, ma_her + reduction_factor*std_her, color='#2196F3', alpha=0.2)

# For RE
plt.plot(indices, ma_re, label='RE', color='#4CAF50', marker='x', linewidth=line_thickness, markevery=global_marker_frequency)
plt.fill_between(indices, ma_re - reduction_factor*std_re, ma_re + reduction_factor*std_re, color='#4CAF50', alpha=0.2)

# For SRE_NC
plt.plot(indices, ma_sre_nc, label='SRE_NC', color='#F44336', marker='s', linewidth=line_thickness, markevery=global_marker_frequency)
plt.fill_between(indices, ma_sre_nc - reduction_factor*std_sre_nc, ma_sre_nc + reduction_factor*std_sre_nc, color='#F44336', alpha=0.2)

# For SRE_NITR
plt.plot(indices, ma_sre_nitr, label='SRE_NITR', color='#9C27B0', marker='^', linewidth=line_thickness, markevery=global_marker_frequency)
plt.fill_between(indices, ma_sre_nitr - reduction_factor*std_sre_nitr, ma_sre_nitr + reduction_factor*std_sre_nitr, color='#9C27B0', alpha=0.2)

# For SRE
plt.plot(indices, ma_sre, label='SRE', color='#FF9800', marker='d', linewidth=line_thickness, markevery=global_marker_frequency)
plt.fill_between(indices, ma_sre - reduction_factor*std_sre, ma_sre + reduction_factor*std_sre, color='#FF9800', alpha=0.2)

# Define text sizes
title_size = 20
label_size = 20
tick_size = 20
legend_size = 15

# Increase tick label size
plt.xticks(fontsize=tick_size)
plt.yticks(fontsize=tick_size)


# Adding title, labels, and legend
plt.title('15x15 Grid Results',fontsize = title_size)
plt.xlabel('# of Episodes',fontsize = label_size)
plt.ylabel('Average Success Rate',fontsize = label_size )
plt.legend(loc='upper left',fontsize = legend_size)  # Shift legend to upper left for better visibility

# Displaying the plot
plt.savefig("15grid.png", format="png", dpi=300)  # Save the plot with 300 dpi resolution
plt.show()
```
